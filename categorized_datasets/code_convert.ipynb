### Functions

```python
def ask_gpt(system, prompt, model):
  return generate(model = model,
        system = system,
        query = prompt,
        temperature=0,
        lastk=0,
        session_id='new',
        rag_usage = True,
        rag_threshold = 0,
        rag_k = 0)
```

#### Convert Blunt

```python
with open("blunt_non_aut.json", "r") as f:
    data = json.load(f)
```

```python
system = '''You are rewriting dialogue responses to adjust interpersonal tone.'''
prompt = '''Your task is to rewrite Speaker 2’s response so that it is less blunt and uses a softer, more socially indirect tone. 

Guidelines:
- Preserve the meaning, intent, and stance of the original response.
- Keep the response natural and conversational.
- Keep length close to the original (slightly longer is fine if needed).
- Make the response significantly less blunt and more socially softened.
- Avoid direct or absolute statements where possible.
- Frame opinions as personal preferences rather than objective judgments.
- Use hedging (e.g., “I think,” “I’m not really,” “kind of,” “a bit”).
- Reduce harsh or evaluative wording (e.g., “bad” → “not really my preference”).
- Where appropriate, slightly cushion the response using indirect phrasing.


Here is the dialogue (only rewrite and return Speaker 2's response as a string):
\n\n{dialogue}'''

for index in range(len(data)):
    dialogue = '''Speaker 1: {statement1}\nSpeaker 2: {statement2}'''
    dialogue = dialogue.format(statement1 = data[index]['Speaker 1'], statement2 = data[index]['Speaker 2'])
    ret = ask_gpt(system, prompt.format(dialogue = dialogue), 'gpt-4o')['response']
    data[index]['Speaker 2']=ret

    
```

```python
with open("blunt_non_aut.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```
