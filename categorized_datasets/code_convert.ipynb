#### Convert Blunt

```python
with open("blunt_non_aut.json", "r") as f:
    data = json.load(f)
```

```python
system = '''You are rewriting responses in a dialogue.'''
prompt = '''Your task is to rewrite Speaker 2’s response so that it is less blunt and uses a softer, more socially indirect tone. 

Guidelines:
- Preserve the meaning, intent, and stance of the original response.
- Keep the response natural and conversational.
- Keep length close to the original (slightly longer is fine if needed).
- Make the response significantly less blunt and more socially softened.
- Avoid direct or absolute statements where possible.
- Frame opinions as personal preferences rather than objective judgments.
- Use hedging (e.g., “I think,” “I’m not really,” “kind of,” “a bit”).
- Reduce harsh or evaluative wording (e.g., change “bad” to “not really my preference”).
- Where appropriate, slightly cushion the response using indirect phrasing.


Here is the dialogue (only rewrite and return Speaker 2's response as a string):
\n\n{dialogue}'''

for index in range(len(data)):
    dialogue = '''Speaker 1: {statement1}\nSpeaker 2: {statement2}'''
    dialogue = dialogue.format(statement1 = data[index]['Speaker 1'], statement2 = data[index]['Speaker 2'])
    ret = ask_gpt(system, prompt.format(dialogue = dialogue), 'gpt-4o')['response']
    data[index]['Speaker 2']=ret

    
```

```python
with open("blunt_non_aut.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```

#### Convert Indirect

```python
with open("indirect_non_aut.json", "r") as f:
    data = json.load(f)
```

```python
system = '''You are rewriting responses in a dialogue.'''

prompt = '''Your task is to rewrite Speaker 2’s response so that it responds to Speaker 1’s indirect speech act/request and provides the information being asked for.

Guidelines:
- Preserve the intent (e.g., agreement or refusal) of the original response.
- Interpret Speaker 1’s underlying intent (e.g., a request, suggestion, or need).
- If the original response is minimal, briefly expand (i.e., don't give too many details but say something beyond yes or no) it into a more helpful response that gives relevant information.
- For example, if Speaker 1 asks “Do you know where the nearest coffee shop is?” and Speaker 2 responds “Yes, I do,” you can rewrite it as “Yes, it’s about two blocks away”. 
- Do NOT create overly detailed responses like “Yes, it is just around the corner. Go forward, turn left, then go right.”
- Keep the response natural and conversational.
- Keep length close to the original.
- Do not use dashes (e.g., - or --).

Here is the dialogue (only rewrite and return Speaker 2's response as a string):
\n\n{dialogue}'''

for index in range(len(data)):
    dialogue = '''Speaker 1: {statement1}\nSpeaker 2: {statement2}'''
    dialogue = dialogue.format(statement1 = data[index]['Speaker 1'], statement2 = data[index]['Speaker 2'])
    ret = ask_gpt(system, prompt.format(dialogue = dialogue), 'gpt-4o')['response']
    data[index]['Speaker 2']=ret
```

```python
with open("indirect_non_aut.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```
