# COCA Frequency Lookup — Figurative Language Dataset

This notebook:
1. Reads all idioms (`fig` field) from `figurative_aut_orig.json` in their **original order**
2. Generates a direct **COCA search URL** for each idiom
3. Prints a numbered list so you can look up each frequency on COCA
4. Creates `fig_data_freq.json` in the same `orig/` folder with `fig` and `freq` fields



```python
import json
import os
from urllib.parse import quote
```

## Verify File Paths

```python
INPUT_FILE  = "figurative_aut_orig.json"
OUTPUT_FILE = "fig_coca_freq.json"

if not os.path.exists(INPUT_FILE):
    raise FileNotFoundError(
        f"Could not find: {INPUT_FILE}\n"
        "Make sure this notebook is placed in the same folder as 'figurative_aut_orig.json'."
    )

print(f"✓ Input file found: {INPUT_FILE}")
```

## Load the JSON and Extract `fig` Values

Idioms are extracted in the **exact same order** as they appear in the original file.


```python
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

fig_list = []
for entry in data:
    fig = entry.get("fig", "").strip()
    if fig:
        fig_list.append(fig)

print(f"✓ Total idioms extracted: {len(fig_list)}")
print()
print("First 5 idioms (order check):")
for i, fig in enumerate(fig_list[:5], 1):
    print(f"  [{i}] {fig}")
```

## Generate COCA Search URLs

Each URL searches COCA for the **exact phrase** (phrase search using quotes).  
Open each URL in your browser while logged into COCA and note the total frequency shown.


```python
COCA_BASE = "https://www.english-corpora.org/coca/?c=coca&q="

def make_coca_url(phrase: str) -> str:
    """Encode an idiom phrase into a COCA exact-phrase search URL."""
    encoded = quote(f'"{phrase}"')
    return COCA_BASE + encoded

# Print numbered list: idiom → URL
print("=" * 72)
print(f"  COCA FREQUENCY LOOKUP — {len(fig_list)} idioms")
print("=" * 72)
print()

for i, fig in enumerate(fig_list, start=1):
    url = make_coca_url(fig)
    print(f"[{i:>3}]  Idiom : {fig}")
    print(f"        URL   : {url}")
    print()

print("-" * 72)
print(f"Total: {len(fig_list)} idioms")
```

## Create `fig_data_freq.json` with Empty Frequency Values

This creates the output JSON file with `freq` set to `null` for every idiom.  
After looking up each frequency on COCA, fill in the values in this file.


```python
output = [
    {"fig": fig, "freq": None}
    for fig in fig_list
]

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(output, f, indent=2, ensure_ascii=False)

print(f"✓ fig_data_freq.json created at: {OUTPUT_FILE}")
print(f"  Total entries: {len(output)}")
print()
print("Preview (first 5 entries):")
for entry in output[:5]:
    print(f"  {entry}")
```

## Verify Output File

Read the saved file back and confirm entry count and order match the original.


```python
with open(OUTPUT_FILE, "r", encoding="utf-8") as f:
    saved = json.load(f)

print(f"Entries in fig_data_freq.json : {len(saved)}")
print(f"Entries in original file      : {len(fig_list)}")
print(f"Order matches                 : {[e['fig'] for e in saved] == fig_list}")
print()
print("All entries (fig | freq):")
print("-" * 50)
for i, entry in enumerate(saved, 1):
    print(f"[{i:>3}]  {entry['fig']!r:<45}  freq = {entry['freq']}")
```
